## 图像的仿射

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import transform as trans
```

```python
image_size = [224, 224]

img_path = './img/Aaron_Sorkin_0001.jpg'
img = cv2.imread(img_path)

print("source image size:{}".format(img.shape))

src = np.array([[85.49595008, 85.73911040000002 + 30],
                [168.39541376, 84.30362432000001 + 30],
                [127.07756863999998, 136.76120383999998 + 30],
                [90.34349120000002, 174.31525376000002 + 30],
                [166.65585920000004, 172.99233343999998 + 30]])

dst = np.array([[99.982100, 121.389000],
               [148.250000, 118.750000],
               [122.701000, 149.671000],
               [102.586000, 172.091000],
               [150.085000, 168.073000]])
tform = trans.SimilarityTransform()
tform.estimate(dst, src)
M = tform.params[0:2, :]
warped = cv2.warpAffine(img, M, (image_size[0], image_size[1]))

plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(warped)
plt.show()
```
